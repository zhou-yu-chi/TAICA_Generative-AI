## 🦋 Butterfly Curve（蝴蝶曲線）

數學公式（極座標定義）：

$$
r(\theta) = e^{\sin\theta} - 2\cos(4\theta) + \sin^5\!\Big(\frac{2\theta-\pi}{24}\Big)
$$

---

### 📌 公式拆解說明
1. **exp(sinθ)**  
   → 隨 θ 變化的「基底半徑」，像呼吸一樣伸縮。  

2. **-2·cos(4θ)**  
   → 四倍速的擺動，製造對稱多瓣翅膀。  

3. **sin⁵((2θ-π)/24)**  
   → 超慢速調變，在區間 [0, 24π] 完成一圈，五次方讓峰谷更尖銳。  

---

✨ 最終圖形呈現多瓣對稱，形似蝴蝶展翅的優美曲線。


```python
plt.style.available   # 回傳可用風格清單
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#套用ggplot風格
plt.style.use("ggplot")

#chatgpt幫忙
# 角度參數：0 → 24π（取樣夠多，曲線更平滑）
theta = np.linspace(0, 24*np.pi, 20000)
# 蝴蝶曲線半徑 r(θ)
r = np.exp(np.sin(theta)) - 2*np.cos(4*theta) + np.sin((2*theta - np.pi)/24)**5
# 直角座標
x, y = r*np.cos(theta), r*np.sin(theta)


#建立一個7*7正方形圖表
plt.figure(figsize=(7,7))

#linewidth=1 線比較細緻
plt.plot(x, y, linewidth=1)

#坐標軸比例相同，避免拉伸變形；並關掉座標軸標記讓圖更乾淨
plt.axis('equal'); plt.axis('off')

#標題
plt.title("Butterfly Curve")
plt.show()
```
